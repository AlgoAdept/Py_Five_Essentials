# Matplotlib tutorial

```python
# It's an open source drawing library which supports rich drawing types.
# It is used to draw 2D and 3D graphics.

# Types Of Plots :
#    1.Bar Chart
#    2.Scatter plot
#    3.Pie Chart
#    4.Histogram
#    5.Line Chart
#    6.Area Graph
```

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
# use numpy to  generate random data

import numpy as np

x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
```

```python
print(np.array([x, y]).reshape(25, 2))
```

```python
# it only takes one command to draw 
pylab.plot(x,y, 'r') # 'r'stands for red
```

```python
# drawing a subgraph

pylab.subplot(1,2,1) # contents of bracket holds (rows, columns ,indexes)
pylab.plot(x,y, 'b--')

pylab.subplot(2,2,2)
pylab.plot(y,x, 'g*-')
```

```python
# (1,2,1) and (1,2,2) are significant if we want tiled subplots ,without which both the graphs will in the same plot 

pylab.subplot(1,2,1)
pylab.plot(x,y, 'r--')
pylab.subplot(1,2,1)
pylab.plot(y,x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5,0.1,0.8,0.8]) # control the left, right, width, height of the canvas from (0 to 1)

axis.plot(x,y, 'r')
```

```python
# again we can draw subgraphs

fig, axes = plt.subplots(nrows = 2, ncols = 1)
for ax in axes: 
    ax.plot(x,y,'y')
```

```python
# we can also draw a picture, or graph inside another graph
fig = plt.figure()

# control the left, right, width, height of the canvas (from 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # small canvas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig = plt.figure(figsize = (16,9), dpi = 600) # New graphic object 
fig.add_subplot()
plt.plot(x,y, 'b')
```

```python
fig = plt.figure(figsize = (16,9), dpi = 50) # New graphic object 
fig.add_subplot()
plt.plot(x,y, 'b')
```

```python
ax.legend(["label1","label2"])

fig, axes = plt.subplots(dpi = 200)

axes.set_title("Title")

axes.set_xlabel("x")
axes.set_ylabel("y")

axes.plot(x, x**2)
axes.plot(x, x**3)

axes.legend(["y = x**2", "y = x**3"],loc=2)
```

```python
# in matplotlib you can set other properties like line color, transparency and more 

fig, axes = plt.subplots(dpi = 150)
axes.plot(x, x**2 ,color = 'red' ,alpha=.9)
axes.plot(x, x**3 ,color = 'red', alpha=.5)
axes.plot(x, x**4, color = 'red' , alpha=.2)
```

```python
fig, ax = plt.subplots(dpi=100)

ax.plot(x, x+1, color='blue', linewidth = 2 ,linestyle='-.')
ax.plot(x, x+3, color='blue', lw = 1.5)
ax.plot(x, x+5, color='blue', linewidth = 2, linestyle = ':')
ax.plot(x, x+7, color='blue', linewidth = 0.5)

line, = ax.plot(x,x+4, color='black' ,lw=1.50)
line.set_dashes([5,10,15,3])

line, = ax.plot(x,x+10, color='red' ,lw=1.50)
line.set_dashes([5,10,4,30])
```

```python
fig, ax = plt.subplots(dpi = 100)

ax.plot(x,x+1, color='blue', marker='o')
ax.plot(x,x+5, color='blue', marker='+')
ax.plot(x,x+3, color='blue', marker='s', lw = 0.5)
ax.plot(x,x+7, color='blue', marker='1')
ax.plot(x,x+9, color='blue', marker='o', markersize = 9, markerfacecolor = 'red')
```
